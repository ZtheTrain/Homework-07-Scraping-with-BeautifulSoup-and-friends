```python
# Running code across all 14 pages of results

base_url = "https://www.tnwb.uscourts.gov/Search/search.aspx"
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
                  "AppleWebKit/537.36 (KHTML, like Gecko) "
                  "Chrome/114.0.0.0 Safari/537.36",
    "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,*/*;q=0.8",
    "Accept-Language": "en-US,en;q=0.9",
}

params = {
    "zoom_sort": 0,
    "zoom_xml": 0,
    "zoom_query": '"CAR"',
    "zoom_per_page": 10,
    "zoom_and": 1,
    "zoom_cat[]": -1
}

all_results = []

for page in range(1, 15):  # Pages 1 to 14
    print(f"\nFetching page {page}...")
    params["zoom_page"] = page
    response = requests.get(base_url, headers=headers, params=params)
    soup = BeautifulSoup(response.text, "html.parser")

    results = soup.find_all(
        "div",
        class_=lambda c: c in ["result_block", "result_altblock"]
    )
    print(f"  Found {len(results)} results")

    for res in results:
        title_div = res.find("div", class_="result_title")
        link_tag = title_div.find("a")
        title = link_tag.get_text(strip=True)
        link = link_tag["href"]

        category_span = title_div.find("span", class_="category")
        category = category_span.get_text(strip=True) if category_span else "N/A"

        context = res.find("div", class_="context")
        context_text = context.get_text(strip=True) if context else ""

        infoline = res.find("div", class_="infoline")
        infoline_text = infoline.get_text(strip=True) if infoline else ""

        all_results.append({
            "Title": title,
            "URL": link,
            "Category": category,
            "Context": context_text,
            "Info": infoline_text
        })

        print(f"  - {title} ({category})")
```

```python
import csv

with open("car_search_results.csv", "w", newline='', encoding="utf-8") as f:
    writer = csv.DictWriter(f, fieldnames=["Title", "URL", "Category", "Context", "Info"])
    writer.writeheader()
    writer.writerows(all_results)

print("\n✅ Done! Results saved to car_search_results.csv")
```

```python
# To download all PDFs

import os
from urllib.parse import urlparse

# Create a directory to save PDFs
os.makedirs("pdfs", exist_ok=True)

for i, result in enumerate(all_results, start=1):
    url = result["URL"].split("#")[0]  # Strip off #search=...
    filename = os.path.basename(urlparse(url).path)

    try:
        print(f"Downloading {filename}...")
        pdf_response = requests.get(url, headers=headers)
        with open(f"pdfs/{filename}", "wb") as f:
            f.write(pdf_response.content)
    except Exception as e:
        print(f"❌ Failed to download {url}: {e}")
```

```python
pdf_folder = "pdfs"
pdf_files = [f for f in os.listdir(pdf_folder) if f.endswith(".pdf")]
print(f"Found {len(pdf_files)} PDF files in '{pdf_folder}'")
```
