```python
dfs = pd.read_html("https://en.wikipedia.org/wiki/List_of_presidents_of_the_United_States", header=0)
dfs
```

```python
len(dfs)
```

```python
combined_df = pd.concat(dfs, axis=1)
combined_df
```

```python
combined_df.info()
```

Only want first 8 columns as the rest is garbage.

```python
combined_df = combined_df.iloc[:, :-10]
```

```python
combined_df.info()
```

```python
combined_df.to_csv("US_presidents-cleaned.csv", index=False)
```
