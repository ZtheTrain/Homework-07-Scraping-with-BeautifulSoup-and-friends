Scrape the list of third party drivers license locations from https://travel-id-documents.az.gov/authorized-third-party-driver-license-locations but include the link. Save as a CSV.

```python
import requests
from bs4 import BeautifulSoup

response = requests.get('https://travel-id-documents.az.gov/authorized-third-party-driver-license-locations')
doc = BeautifulSoup(response.text, 'html.parser')
```

```python
tds = doc.find_all('td')
tds
```

```python
# Skip the first 4 header entries if present
if tds[0].text.strip().lower() == "company":
    tds = tds[4:]

# Group every 4 <td> tags as one record
records = []
for i in range(0, len(tds), 4):
    chunk = tds[i:i+4]
    if len(chunk) == 4:
        company_td = chunk[0]

        # Text of company
        company_name = company_td.get_text(separator=" ", strip=True)

        # Check for <a> tag inside company <td>
        a_tag = company_td.find('a')
        company_link = a_tag['href'] if a_tag else None

        address = chunk[1].get_text(separator=" ", strip=True)
        telephone = chunk[2].get_text(separator=" ", strip=True)
        hours = chunk[3].get_text(separator=" ", strip=True)

        records.append({
            'Company': company_name,
            'Company URL': company_link,
            'Address': address,
            'Telephone': telephone,
            'Hours': hours
        })

# Example output
for r in records[:3]:
    print(r)
```

```python
import pandas as pd

df = pd.DataFrame(records)
df
```

```python
df.to_csv("AZ_third_party_drivers_license_locations.csv", index=False)
```
